```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
# read our dataset as dataframe 
data=pd.read_csv("dataset/Cancer_Data.csv")
#first five rows 
data.head()
```

```python
# what about some information about the data
data.info()
```

```python
data=data.drop(columns=["Unnamed: 32"])
```

```python
data
```

```python
#let's see if we have any duplicated rows 
data.duplicated().sum()
```

we donot have any duplicated rows let's focus on columns 


```python
# the column id is not important let's drop it 
data.drop(columns=["id"],inplace=True)
```

```python
data.columns
```

```python
# scaling the columns 
from sklearn.preprocessing import MinMaxScaler
# Initialize MinMaxScaler
scaler = MinMaxScaler()
#columns we want to scale 
scaled_columns=['radius_mean','texture_mean','perimeter_mean','area_mean','radius_worst','perimeter_se','area_se','radius_worst','perimeter_worst','area_worst']
data[scaled_columns]=scaler.fit_transform(data[scaled_columns])
```

```python
#now let's see the scaled data
data
```

```python
#our data is now ready let's build our model
features=data.drop(columns=['diagnosis'])
trarget=data['diagnosis']
```

```python
features
```

```python
trarget
```

```python
# now let's encode the target =>convert it from categorical to numerical
from sklearn.preprocessing import LabelEncoder
label_encode=LabelEncoder()
target_encoded=label_encode.fit_transform(trarget)
target_encoded
```

```python
# reshaping the target to 2_d array
target = np.reshape(target_encoded, (-1, 1))
target
```

```python
#split the data into train and test
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(features,target,test_size=0.3,random_state=42)
```

```python
import tensorflow as tf 
#building the model
model=tf.keras.Sequential([
    tf.keras.layers.Dense(64,activation='relu',input_shape=(30,)),
    tf.keras.layers.Dense(32,activation='relu'),
    tf.keras.layers.Dense(1,activation='sigmoid')# Output layer with sigmoid activation for binary classification

])
```

```python
# Compiling the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
# Training the model
history = model.fit(X_train, y_train, epochs=10, batch_size=32, validation_data=(X_test, y_test))
```

```python
# Evaluating the model
test_loss, test_accuracy = model.evaluate(X_test, y_test)
print('test Loss:',test_loss)
print('test Accuracy:',test_accuracy)
```
